```python
import pandas as pd
import re
from openpyxl import load_workbook
```

```python
pd.set_option('display.max_rows', 100)
```

```python
census2011provlocalmun = pd.read_csv("census-2011-2016-boundaries-pr-mn-highest-edu-level-20-24.csv", skiprows=9)
census2011districtmun = pd.read_csv("census-2011-2016-boundaries-dc-highest-edu-level-20-24.csv", skiprows=9)
census2011 = pd.concat([census2011provlocalmun, census2011districtmun])
census2011
```

```python
census2011 = census2011.drop(columns=["Summation Options", "Age in completed years", "Unnamed: 5"])
census2011.columns = ["Geography", "Highest level of education", "Count"]
census2011 = census2011[census2011["Geography"] != "Total"]
census2011 = census2011[census2011["Highest level of education"] != "Total"]
census2011 = census2011.dropna()
census2011
```

```python
commsurv2016 = pd.read_csv("community-survey-2016-pr-dc-mn-highest-edu-level-20-24.csv", skiprows=9)
commsurv2016
```

```python
commsurv2016 = commsurv2016.drop(columns=["Summation Options", "Age - 5 year age groups", "Unnamed: 5"])
commsurv2016.columns = ["Geography", "Highest level of education", "Count"]
commsurv2016 = commsurv2016[commsurv2016["Geography"] != "Total"]
commsurv2016 = commsurv2016[commsurv2016["Highest level of education"] != "Total"]
commsurv2016 = commsurv2016.dropna()
commsurv2016
```

```python
replacements = [
    ("(CS2016) Any tertiary", "Any tertiary"),
    ("(CS2016) Grade 10/11 equivalent", "Grade 10/11 equivalent"),
    ("(CS2016) Grade 9", "Grade 9"),
    ("(CS2016) Less than Grade 9", "Less than Grade 9"),
    ("(CS2016) Matric/Metric equivalent", "Matric/Matric equivalent"),
]
for (old, new) in replacements:
    commsurv2016["Highest level of education"].replace(old, new, inplace=True)
```

```python
census2011["Geography"] = census2011["Geography"].apply(lambda cell: re.sub(" *:.+", "", cell))
census2011 = census2011.drop_duplicates(["Geography", "Highest level of education"])
```

```python
commsurv2016["Geography"] = commsurv2016["Geography"].apply(lambda cell: re.sub(" *:.+", "", cell))
commsurv2016 = commsurv2016.drop_duplicates(["Geography", "Highest level of education"])
```

```python
highestedu2011_2016 = census2011.merge(commsurv2016, how="right", on=["Geography", "Highest level of education"], suffixes=("_2011", "_2016"))
```

```python
replacements = [
    ("Eastern Cape", "EC"),
    ("Northern Cape", "NC"),
    ("Western Cape", "WC"),
    ("Gauteng", "GT"),
    ("Limpopo", "LIM"),
    ("Mpumalanga", "MP"),
    ("KwaZulu-Natal", "KZN"),
    ("Free State", "FS"),
    ("North West", "NW"),
]
for (prov, code) in replacements:
    highestedu2011_2016["Geography"].replace(prov, code, inplace=True)
```

```python
highestedu2011_2016.loc[highestedu2011_2016["Geography"] == "CPT"] 
```

```python
highestedu2011_2016_grouped = highestedu2011_2016.groupby(["Geography", "Highest level of education"])
highestedu2011_2016_grouped = highestedu2011_2016_grouped.sum(numeric_only=True)

# highestedu_abs_change represents the absolute change in population of a given education level group in a given geography.
highestedu2011_2016_grouped["highestedu_abs_change"] = highestedu2011_2016_grouped.apply(lambda row: row["Count_2016"] - row.Count_2011, axis = 1)

# 2011_highestedu_proportion represents the proportion of an aeducation level group of the total population for a geography in 2011
highestedu2011_2016_grouped["2011_highestedu_proportion"] = highestedu2011_2016_grouped['Count_2011'] / highestedu2011_2016_grouped.groupby('Geography')['Count_2011'].transform('sum')
# 2016_highestedu_proportion represents the proportion of an education level group of the total population for a geography in 2016
highestedu2011_2016_grouped["2016_highestedu_proportion"] = highestedu2011_2016_grouped['Count_2016'] / highestedu2011_2016_grouped.groupby('Geography')['Count_2016'].transform('sum')

# highestedu_proportion_change represents the change in the proportion an education level group represents
# as a total of the population in that geography. e.g. if the education level group "Grade 9 represented 
# 20% of the population in 2011 and 25% of the population in 2016, then highestedu_proportion_change 
# for education level group "Grade 9" would be 5 in that geography.
highestedu2011_2016_grouped["highestedu_proportion_change"] = highestedu2011_2016_grouped["2016_highestedu_proportion"] - highestedu2011_2016_grouped["2011_highestedu_proportion"]

# highestedu_proportion_change_pct represents the percentage by which the proportion of each 
# education level group in the geography changed. Since the absolute change in proportion is likely to 
# be proportional to the original proportion, this variable is more useful to highlight changes
# that are significant relative to the original population.
highestedu2011_2016_grouped["highestedu_proportion_change_pct"] = highestedu2011_2016_grouped["highestedu_proportion_change"] / highestedu2011_2016_grouped["2011_highestedu_proportion"]

# highestedu_magnitude_pct_change represents the percentage that an education level group changed. 
# i.e. if the education level group "Grade 9 had 100,000 people, and now it has 120,0000 people,
# its highestedu_magnitude_pct_change value would be 0.20, i.e. the "Grade 9" education level group grew by 20%.
highestedu2011_2016_grouped["highestedu_magnitude_pct_change"] = highestedu2011_2016_grouped["highestedu_abs_change"] / highestedu2011_2016_grouped["Count_2011"]
highestedu2011_2016_grouped
```

```python
highestedu2011 = highestedu2011_2016_grouped[["Count_2011"]].copy()
highestedu2011["year"] = "2011"
highestedu2011.columns = ["Count", "year"]
highestedu2016 = highestedu2011_2016_grouped[["Count_2016"]].copy()
highestedu2016["year"] = "2016"
highestedu2016.columns = ["Count", "year"]
filename = "highestedu2011_2016_year_var.xlsx"
pd.concat([highestedu2011, highestedu2016]).reset_index().to_excel(filename, index=False)
wb = load_workbook(filename)
ws = wb.create_sheet(title="Documentation")
ws["A1"] = "Highest level of education attained by 20-24-year olds at 2011 and 2016."
wb.save(filename)
```

```python
highestedu2011_2016_abs_change = highestedu2011_2016_grouped[["highestedu_abs_change"]].copy()
highestedu2011_2016_abs_change.columns = ["Count"]
filename = "highestedu2011_2016_abs_change.xlsx"
highestedu2011_2016_abs_change.reset_index().to_excel(filename, index=False)
wb = load_workbook(filename)
ws = wb.create_sheet(title="Documentation")
ws["A1"] = "Absolute population change between 2011 and 2016 for each age group in each geography."
wb.save(filename)
```

```python
highestedu2011_2016_proportion_change_pct = highestedu2011_2016_grouped[["highestedu_proportion_change_pct"]].copy()
highestedu2011_2016_proportion_change_pct.columns = ["Count"]
filename = "highestedu2011_2016_proportion_change_pct.xlsx"
highestedu2011_2016_proportion_change_pct.reset_index().to_excel(filename, index=False)
wb = load_workbook(filename)
ws = wb.create_sheet(title="Documentation")
ws["A1"] = (
    "This represents the percentage by which the proportion of each\n"
    "age group in the geography changed. Since the absolute change in proportion is likely to\n"
    "be proportional to the original population proportion of the age group, this variable is more useful to highlight changes\n"
    "that are significant relative to the original population."
)
wb.save(filename)
```

```python
highestedu2011_2016_magnitude_pct_change = highestedu2011_2016_grouped[["highestedu_magnitude_pct_change"]].copy()
highestedu2011_2016_magnitude_pct_change.columns = ["Count"]
filename = "highestedu2011_2016_magnitude_pct_change.xlsx"
highestedu2011_2016_magnitude_pct_change.reset_index().to_excel(filename, index=False)
wb = load_workbook(filename)
ws = wb.create_sheet(title="Documentation")
ws["A1"] = (
    "This represents the percentage that an age group population changed between 2011 and 2016.\n"
    "i.e. if the age group 0-14 had 100,000 people, and now it has 120,0000 people,\n"
    "its highestedu_magnitude_pct_change value would be 0.20, i.e. the 0-14 age group grew by 20%."
)
wb.save(filename)
```
